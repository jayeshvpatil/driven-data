```python
for ds in combine:
    median_const = ds['construction_year'].median()
    ds['construction_year'] = ds.construction_year.mask(ds.construction_year ==0, median_const).astype(int)
    median_gpsh = ds['gps_height'].median()
    ds['gps_height'] = ds.gps_height.mask(ds.gps_height ==0, median_const).astype(int)
    ds['date_recorded'] = pd.to_datetime(ds['date_recorded']).dt.year
    ds['date_recorded'] = ds['date_recorded'].astype('int32')
    ds['date_recorded'] = ds['date_recorded']-ds['construction_year']
    
    

    
```

```python
#train_df['status_group']=train_df['status_group'].astype('category')
#train_df['status_group']= train_df['status_group'].cat.codes
train_df['status_group'] = train_df['status_group'].map( {'functional': 0, 'non functional': 1, 'functional needs repair':2} ).astype(int)
#train_df.head(10)
```

```python
print("Before", train_df.shape, test_df.shape, combine[0].shape, combine[1].shape)

train_df = train_df.drop(['region','num_private','wpt_name','recorded_by','scheme_name'], axis=1)
test_df = test_df.drop(['region','num_private','wpt_name','recorded_by','scheme_name'], axis=1)
combine = [train_df, test_df]

"After", train_df.shape, test_df.shape, combine[0].shape, combine[1].shape
```

```python

for ds in combine:
    ds['basin']= pd.get_dummies(ds['basin'])
    ds['lga']= pd.get_dummies(ds['lga'])
    ds['ward']= pd.get_dummies(ds['ward'])
    ds['public_meeting'].fillna(False)
    ds['public_meeting']= pd.get_dummies(ds['public_meeting'])
    ds['permit']= pd.get_dummies(ds['permit'])
    ds['management']= pd.get_dummies(ds['management'])
    ds['scheme_management'].fillna('-999', inplace=True)
    ds['scheme_management']= pd.get_dummies(ds['scheme_management'])
    ds['management_group']= pd.get_dummies(ds['management_group'])
    ds['extraction_type_class']= pd.get_dummies(ds['extraction_type_class'])
    ds['payment_type']= pd.get_dummies(ds['payment_type'])
    ds['quantity']= pd.get_dummies(ds['quantity'])
    ds['source']= pd.get_dummies(ds['source'])
    ds['waterpoint_type']= pd.get_dummies(ds['waterpoint_type'])
    ds['water_quality']= pd.get_dummies(ds['water_quality'])
    ds['waterpoint_type']= pd.get_dummies(ds['waterpoint_type'])
    ds['waterpoint_type_group']= pd.get_dummies(ds['waterpoint_type_group'])
    ds['quality_group']= pd.get_dummies(ds['quality_group'])
    ds['source_type']= pd.get_dummies(ds['source_type'])
    ds['source_class']= pd.get_dummies(ds['source_class'])
    ds['quantity_group']= pd.get_dummies(ds['quantity_group'])
    ds['payment']= pd.get_dummies(ds['payment'])
    ds['extraction_type']= pd.get_dummies(ds['extraction_type'])
    ds['extraction_type_group']= pd.get_dummies(ds['extraction_type_group'])
    ds['funder'].fillna('-999', inplace=True)
    ds['funder']= pd.get_dummies(ds['funder'])
    ds['subvillage']= pd.get_dummies(ds['subvillage'])
    ds['installer'].fillna('-999', inplace=True)
    ds['installer']= pd.get_dummies(ds['installer'])
    ds['population'].replace(0,train_df.groupby(['lga','district_code'])["population"].transform("mean"), inplace=True)
    
    
    

  
train_df.head()
```

```python
from sklearn.model_selection import train_test_split

X= train_df.drop(['id','status_group'],axis=1)
y= train_df["status_group"]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=22)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc = sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)

X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=400)
random_forest.fit(X_train, y_train)
Y_pred_winner = random_forest.predict(X_test)
random_forest.score(X_train, y_train)
acc_random_forest = round(random_forest.score(X_train, y_train) * 100, 2)
acc_random_forest
```

```python
holdout= test_df.drop(['id'],axis=1)
holdout_scaled = sc.transform(holdout)
holdout_pred = random_forest.predict(holdout_scaled)

print(holdout_pred)
```

```python
submission = pd.DataFrame({
        "id": test_df["id"],
        "status_group": holdout_pred
    \
    })
submission['status_group']=submission['status_group'].map( { 0: 'functional',1: 'non functional', 2:'functional needs repair'} ).astype(str)

submission.to_csv('submission.csv', index=False)
```
