```python

combine =[X_train, X_test]
for ds in combine:
    ds['region_code'] = ds['region_code'].astype(object)
    ds['district_code'] = ds['district_code'].astype(object)

    # Fix construction year =0 and replace with median. 
    # Also calculate age of the pump and drop the daterecorded & construction year
    ds['construction_year']= pd.to_datetime(ds['construction_year'].astype(int)).dt.year
    median_const = ds['construction_year'].median()
    ds['construction_year'] = ds.construction_year.mask(ds.construction_year ==0, median_const).astype(int)
    ds['date_recorded'] = pd.to_datetime(ds['date_recorded']).dt.year
    ds['date_recorded'] = ds['date_recorded'].astype('int32')
    ds['age'] = ds['date_recorded']-ds['construction_year']
    ds.drop(['construction_year','date_recorded'],inplace=True, axis=1)
    #  Fix GPS_height =0 and replace with median
    median_gpsh = ds['gps_height'].median()
    ds['gps_height'] = ds.gps_height.mask(ds.gps_height ==0, median_const).astype(int)
    # missing population..fix by mean for similar district
    ds['population'].replace(0,ds.groupby(['lga','district_code'])["population"].transform("mean"), inplace=True)
```
